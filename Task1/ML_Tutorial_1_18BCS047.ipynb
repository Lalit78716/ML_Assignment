```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu,sigma=5,1
x = np.random.normal(mu,sigma,10)
x
```

```python
np.mean(x)
```

```python
y = 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-0.5*((x-mu)/sigma)**2)
y
```

```python
sns.lineplot(x,y,color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig("guassian.png")
```

```python
space=np.linspace(1,10,12)
space
```

```python
def liklihood(z):
    ans= 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-0.5*((x-z)/sigma)**2)
    return ans
```

```python
plt.figure(figsize = (20,15))
for i in range(1,12,1):
    y1=liklihood(space[i])
    plt.subplot(3,4,i)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('mean='+str(round(space[i],6))+', variance='+str(round(space[i],6)),color='blue')
    sns.lineplot(x,y1,color='red')
    plt.savefig("Subplot.png")
    
```
