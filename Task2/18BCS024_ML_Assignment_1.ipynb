```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm,stats
```

```python
import numpy as np
```

```python
import math
mu = 5
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 1.60*sigma, mu + 1.60*sigma,10)
y=norm.pdf(x,mu,sigma)
plt.plot(x,y,'-o',color='b',linestyle='-')
plt.title('Gaussian Distribution of 10 samples with μ = 5 and σ = 1')
#plt.show()
plt.savefig('gaussian.png')
```
